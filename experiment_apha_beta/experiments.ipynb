```python
penalty_dict = {}
for elem in penalty_list:
    nelem = (elem[0][0],elem[0][1],elem[0][2])
    if(nelem not in penalty_dict):
        penalty_dict[nelem] = []
    penalty_dict[nelem].append(elem[1])
```

```python
penalty_avg = {}
for elem in penalty_dict:
    penalty_avg[elem] = sum(penalty_dict[elem])/5
```

```python
def final_results(gs,N,a_salary,c_salary,f,R):
    ans_list = []
    for elem in penalty_avg:
        if(elem[0]!=gs or elem[1]!=N):
            continue
        else:
            a_pop = int(N*(elem[2]/100))
            c_pop = N - a_pop
            suma_salary = a_pop*math.log(a_salary)
            sumc_salary = c_pop*math.log(c_salary)
            suma_penalty = R*penalty_avg[elem]
            try:
                sumc_penalty = (1/f)*(c_pop/a_pop)*(suma_penalty)
            except:
                sumc_penalty = 0
            ans_list.append([elem,suma_salary,suma_penalty,sumc_salary,sumc_penalty])
    ans_list.sort()
    for elem in ans_list:
        print(str(elem[0])+":"+str((int(elem[1]),int(elem[2]),int(elem[3]),int(elem[4]))),end=" = ")
        print((elem[1]+elem[3]-elem[2]-elem[4]))
    with open("./results/"+str((gs,N,a_salary,c_salary,f,R)),'w') as f:
        for item in ans_list:
            f.write(str(item)+" : " +str((item[1]+item[3]-item[2]-item[4])) + "\n")
    return ans_list
```
